#### 关于书写： 写 x , y ,p 或者对应写 input, target, output , 对应称呼 输入样本 目标样本 输出样本

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# 有以下数据集，
    # 但是其生成的张量的尺寸是一维的
x = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]  
y = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0]
x = torch.tensor(x)
y = torch.tensor(y)


x.shape,y.shape
    # 这个尺寸没有样本数和样本大小的说法,对吗？
```

```python
# 对input进行分批处理 => 2D数据： Batch,Nin
    # 在dim = 1 
x = x.unsqueeze(dim=1)
# y = y.unsqueeze(dim=1) #target.shape is must 1D ?? 
x.shape,y.shape
    # 这个尺寸也就是说11个样本，每个样本大小为1(每个样本的特征数为1)
```

```python
# 拆分训练集和验证集

n_samples = x.shape[0]  # 总样本数
n_val = int(n_samples*0.2)  # 验证样本数

shuffled_indices =  torch.randperm(n_samples)  # 洗牌后的索引们 #indices是index的复数
train_indices = shuffled_indices[:-n_val]   
val_indices = shuffled_indices[-n_val:]
    # 这里的切分操作容易出错

x_train = x[train_indices] #训练集
y_train = y[train_indices]

x_val = x[val_indices]  # 验证集
y_val = y[val_indices]

x_train.shape,x_val.shape,y_train.shape,y_val.shape
```

```python
# 归一化（也就是特征放缩）  
    #  在知乎上看见一句话： 验证集和训练集，要归一都归一，要不归一都不归一。
x_trainn = (x_train-x_train.mean())/x_train.std()
x_valn = (x_val-x_val.mean())/x_val.std() 
```

```python
### 测试而已    # 测试神经网络的forward

in_tmp =torch.ones(5,1) # input

input_model = nn.Linear(1,8) # 输入模块
hidden_model = nn.Linear(8,8) # 隐层模块
output_model = nn.Linear(8,1) # 输出模块
activate_fn = torch.nn.ReLU() # 激励函数
    
    # forward
o1 = input_model(in_tmp)  #(5,1)x(1,8) > (5,8)  #输入模块的output
activate_fn(o1)
o2 = hidden_model(o1) #(5,8)x(8,8) > (5,8)
activate_fn(o2)
o3 = output_model(o2) # (5,8)x(8,1) > (5,1)  # 输出模块的output

o3,input_model.weight,input_model.bias
```

```python
# 定义神经网络model
    # 备注：nn.Sequential 是一个时序容器。Modules 会以他们传入的顺序被添加到容器中。当然，也可以传入一个OrderedDict。

from collections import OrderedDict 
    
orderedict = OrderedDict([
    ("input_linear",nn.Linear(1,8)),
    ("input_activate",nn.Tanh()),
    ("hidden_linear",nn.Linear(8,8)),
    ("hidden_activate",nn.ReLU()),
    ("output_linear",nn.Linear(8,1))
])
seq_model = nn.Sequential(orderedict)
    # 个人理解：input(?,1)xinput_model(1,8)xhidden_model(8,8)xoutput_model(8,1) => output(?,1)
        # 除了输出模块，其余的模块之后都要放置激励函数。
    # nn.Linear 适合对输入数据做线性变换：y=Ax+b
    # 注意：nn.Linear()中的参数是 in_feature out_feature bias , 是每个样本的大小，而不是样本个数。 

seq_model,seq_model.output_linear.weight
```

```python
# 定义损失函数

#loss_fn = nn.CrossEntropyLoss() #交叉熵损失函数 #MSELoss()  
def loss_fn(p,y):
    squared_diffs = (p-y)**2  
    return squared_diffs.mean()  
```

```python
# 定义优化器
    # 构造优化器的参数列表由model.parameters()提供
optimizer = optim.Adam(params=seq_model.parameters(),lr=1e-1)
```

```python
# 定义训练循环函数

def train_loop(n_epochs,optimizer,model,loss_fn,
              x_train,y_train,x_val,y_val,
              verbose=500):
    for epoch in range(1,n_epochs+1):
        
        # 损失张量  (由model执行了forward)
        loss_train = loss_fn(model(x_train),y_train.long())    # 参数 input, target
        loss_val = loss_fn(model(x_val),y_val.long()) 
        #
        optimizer.zero_grad()  # 优化器清零grad
        loss_train.backward()  # loss_train执行了backward
        optimizer.step()       # 优化器更新params
        
        if epoch % verbose == 0 :
            print('Epoch %d, Train_loss %.4f, Val_loss %.4f'%
                 (epoch,float(loss_train),float(loss_val)))
        
        if epoch==n_epochs:
            print('output_val',seq_model(x_val))
            print('target_val', y_val)        
        
```

```python
# 调用训练循环函数

train_loop(n_epochs=10000,optimizer=optimizer,model=seq_model,loss_fn=loss_fn,
           x_train=x_trainn,y_train=y_train,x_val=x_valn,y_val=y_val,
           verbose = 500)
```
